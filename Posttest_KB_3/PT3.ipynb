```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid") # Plih Style

df = pd.read_csv("bodyPerformance.csv")
```

```python
df.head()
```

```python
# GRAFIK BATANG
plt.figure(figsize=(10,10)) # panjang 10, lebar 10
sns.countplot(data=df,x= 'class', palette='crest', hue='gender')
plt.title('Perbandingan class kinerja pada gender')
plt.xlabel('Class')
plt.ylabel('Kuantitas')
plt.grid()
plt.show()
```

**Grafik Batang yang menggunakan seaborn yang menjelaskan tentang jumlah data class kinerja tubuh pada gender dari dataset yang berjumlah 3000 data dengan xlabel: class dan ylabel: count**

```python
dp = df["gender"].value_counts().rename_axis("gender").reset_index(name="id_count")
```

```python
# GRAFIK PIE
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

dfGroup = df.groupby("class")['class'].count()
dfGroup1 = df.groupby("gender")['gender'].count()


fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])

fig.add_trace(go.Pie(values=dfGroup,name = "CO2 Emisions",labels = dfGroup.index,hole=.3,),1, 1)
fig.add_trace(go.Pie(values=dfGroup1,name = "CO2 Emisions",labels = dfGroup1.index,hole=.3),1, 2)

fig.update_traces(textposition='inside', textinfo='percent+label')

fig.update_layout(
    title_text="Perbandingan class dan gender",
    annotations=[dict(text='Class', x=0.19, y=0.5, font_size=20, showarrow=False),
                 dict(text='Sex', x=0.80, y=0.5, font_size=20, showarrow=False)],
    showlegend=False)

fig.show()
```

**Grafik Pie ini menampilkan data dalam bentuk Persenan dari Jumlah class dan gender yang berada pada dataset yang totalnya berjumlah 3000**

```python
# GRAFIK HEATMAP
plt.figure(figsize=(25,25))
sns.heatmap(df.corr(),annot=True,cmap='Blues',annot_kws={'size':15})
plt.title('Korelasi antar atribut pada Kinerja tubuh')
plt.show()
```

**Grafik HeatMap ini merupakan visualisasi atau pemetaan dengan menampilkan dari dataset Body Performance pada pria dan wanita, yang menjelaskan korelasi antar variabel yang semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap.**
